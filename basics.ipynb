```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

%matplotlib notebook

fruits  = pd.read_table ("C:/Users/Crysis/PycharmProjects/DataScience/MachineLearning/fruit_data_with_colors.txt")

fruits.head()
```

#We split the dataset into features (X) and labels (y). These two dataframes are then further split into training and test datasets (75:25). Therefore, we get X_train and y_train for training and X_test and y_test for testing.  

```python
X = fruits[['height', 'width' , 'mass', 'color_score']]
y = fruits['fruit_label']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from matplotlib import cm

cmap = cm.get_cmap('gnuplot')
scatter = pd.scatter_matrix(X_train, c= y_train, marker = 'o', s=40, hist_kwds={'bins':15}, figsize=(12,12), cmap=cmap)

plt.show()
```

#3D scatter plot. Don't ask me how it works.

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_train['width'], X_train['height'], X_train['color_score'], c= y_train, marker ='o', s=10)

ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('color_score')
plt.show()
```
