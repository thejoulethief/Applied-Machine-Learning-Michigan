So this is the KNearestNeighbors classifer without normalization. Accuracy seems maximum for K=1, and ranges from 0.6-0.73 

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

fruits = pd.read_table("fruit_data_with_colors.txt")

fruits_features=['mass', 'width', 'height', 'color_score']

X_fruits = fruits[fruits_features]

Y_fruits = fruits['fruit_label']
target_names_fruits = ['apple', 'orange', 'mandarin', 'lemon']


X_train, X_test, y_train, y_test = train_test_split(X_fruits, Y_fruits, random_state=0)

score_list=[]

for k in range(1,20):

    knn = KNeighborsClassifier(n_neighbors=k)

    knn.fit(X_train, y_train)

    score = knn.score(X_test, y_test)

    score_list.append(score)
    print("The score for k value of {} was {}".format(k,score))


plt.scatter(range(1,20), score_list, marker='o')
plt.show()
```

KNearestNeighbor classifier with normalization using MinMax scaling. 

```python

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)

print(X_train_scaled)
#X_train.head()

# Interesting. X_train is a pandas dataframe, while the output of the scaler fucntion is an np array

X_test_scaled = scaler.transform(X_test)

# We would not fit the scaling model to both X_train and X_test, since it would result in a mismatch and random values
# Also, if we train the scaler on the test dataset and not the training dataset, there would be "data leakage"

# Now we retrain the KNearNeighbor classifer with the normalized dataset

score_list = []

for k in range(1,20):
    
    knn = KNeighborsClassifier(n_neighbors=k).fit(X_train_scaled, y_train)

    score = knn.score(X_test_scaled, y_test)
    
    score_list.append(score)
    
    print("K value:{}, Score:{:.2f}".format(k, score))

plt.scatter(range(1,20), score_list, marker='o')
plt.show()
```

So therefore, for KNN classifier, without normalization, the optimal K value is 1 and the score is 0.6
                                While, with normalization, the optimal K value is 5 and the score is 1. 

```python
# Creating a sample binary regression problem dataset
%matplotlib notebook

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap


from sklearn.datasets import make_regression


plt.title("Sample regression dataset")

X_R1, Y_R1 = make_regression(n_samples=100, n_features=1, noise = 30,n_informative=1, bias=150.0, random_state=0)

plt.scatter(X_R1, Y_R1, marker='o',s=50)
plt.show()

from sklearn.neighbors import KNeighborsRegressor


X_train, X_test, y_train, y_test = train_test_split(X_R1, Y_R1, random_state=0)

knn = KNeighborsRegressor(n_neighbors=5)

knn.fit(X_train, y_train)

score_train = knn.score(X_train, y_train)
score_test = knn.score(X_test, y_test)

print(score_train, score_test)

```

Normalization 

```python
%matplotlib notebook
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train, y_train)

X_train_s = scaler.transform(X_train)
X_test_s = scaler.transform(X_test)

from sklearn.neighbors import KNeighborsRegressor

score_highest=0

for k in range(1,50):
    knn = KNeighborsRegressor(n_neighbors=k).fit(X_train_s, y_train)

    score_test = knn.score(X_test_s, y_test)
    score_train = knn.score(X_train_s, y_train)
    
    
    if score_test>score_highest:
        k_value = k
        score_highest=score_test
        

print("K: {}, score: {:.2f}".format(k_value,score_highest))

# Just did this to see the normalization on the X_training set
plt.scatter(X_train_s, y_train, marker='o', s=50)
plt.show()
```

Kind of a contradictory name but logistic regression is used for classification between binary choices or multiclass target values. Since you take on that logictic function with y belongs to (0,1) at the output. 

Applying it to the fruits dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split

fruits = pd.read_table('fruit_data_with_colors.txt')

features = ['height', 'width']
X_fruits_2d = fruits[features]
Y_fruits = fruits['fruit_label']

y_fruits_apple = Y_fruits == 1   # make into a binary problem: apples vs everything else
X_train, X_test, y_train, y_test = (
train_test_split(X_fruits_2d.as_matrix(),
                y_fruits_apple.as_matrix(),
                random_state = 0))


from sklearn.linear_model import LogisticRegression

logr = LogisticRegression(C=100).fit(X_train, y_train)

logr.score(X_test, y_test)

h = int(input('Height of the fruit'))
w = int(input('Width of the fruit'))
          


print('For a height of {} and weight of {}, the prediction is {}'.format(h, w, ['not an apple', 'an apple'][logr.predict([[h,w]])[0]]))

```
