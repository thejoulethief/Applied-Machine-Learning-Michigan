```python
%matplotlib notebook
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

fruits  = pd.read_table ("C:/Users/Crysis/PycharmProjects/DataScience/MachineLearning/fruit_data_with_colors.txt")

fruits.head()
```

```python
X = fruits[['mass', 'width', 'height']]
y = fruits['fruit_label']

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
           
```

```python
lookup_fruit_name = dict(zip(fruits.fruit_label.unique(), fruits.fruit_name.unique()))
```

K_nearest neighbors classifier object

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors =1)
```

Fit training data to the classifier

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

The KNN model predicts a label given parameters of mass, width and height. The numpy array thus returned is fed into the dictionary with labels as keys and names as items that we previously created, which gives us the name of the fruit. 

```python
fruit_prediction = knn.predict([[170, 9, 7.5]])
lookup_fruit_name[fruit_prediction[0]]
```
